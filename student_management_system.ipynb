```python
# full_name, age, grade, id
# fun_req : add, update, delete, read, search, menu
# storage: file, menu
import os
import csv

FILE_NAME ="students.csv"
FIELDS = ["id", "name", "age", "grade"]

def load_data():
    students = []
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, mode="x") as file:
            print("File Created!")
    
    with open(FILE_NAME, mode="r",newline="") as file:
        reader = csv.DictReader(file)
        for row in reader:
            students.append(row)

    return students


def view_student():
    students = load_data()
    if not students:
        print("File is empty!")
        return

    for s in students:
        print(f"ID:{s["id"]}, Name:{s["name"]}, Age:{s["age"]}, Grade:{s["grade"]}")

def save_students(students):
    with open(FILE_NAME, mode="w", new_line="") as file:
        writer  = csv.DictWriter(file, fieldnames=FIELDS)
        writer.writeheader()
        writer.writerows(students)


def add_student():
    students = load_data()
    student_id = input("Enter Student ID: ")
    for  s in students:
        if s["id"] == student_id:
            print("Id must be unique!")
            return 

    name = input("Enter Student Name")    
    age = input("Enter Student Age")    
    grade = input("Enter Student Grade") 

    students.append({
        "id" : student_id,
        "name": name,
        "age" : age,
        "grade": grade
    })   
    
    save_students(students)

def search_student():
    students = load_data()
    student_id = input("Enter Student ID: ")

    for s in students:
        if s["id"] == student_id:
            print(f"ID:{s["id"]}, Name:{s["name"]}, Age:{s["age"]}, Grade:{s["grade"]}")
            return

    print(" Data not found!")


def update_student():
    students = load_data()
    student_id = input("Enter Student ID: ")

    for s in students:
        if s["id"] == student_id:
            s["name"] = input("Enter New Name: ")
            s["age"] = input("Enter New Age: ")
            s["grade"] = input("Enter New Grade: ")
            save_students(students)
            print("Data has been Update!")
            return 

    print("Data not found!")


def delete_student():
    students = load_data()
    student_id = input("Enter Student ID: ")
    new_students = []

    for s in students:
        if s[id] != student_id:
            new_students.append(s)
            
    save_students(new_students)




def menu():
    while True:
        print("********** STUDENT MANAGEMENT SYSTEM ***********")
        print("1. View Student Data")
        print("2. Add Student Data")
        print("3. Update Student Data")
        print("4. Search Student Data")
        print("5. Delete Student Data")
        print("6. Exit App")

        operation = int(input("Choose Opeartion"))

        match operation:
            case 1:
                print(" ******* View *********")
                view_student()
            case 2:
                print("Add")
            case 3:
                print("Update")
            case 4:
                print("Search")
            case 5:
                print("Delete")
            case 6:
                print("Exit")
                break
            case _:
                print("Try Again!")

menu()
```
